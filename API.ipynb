```python
# pip install XRetriever

from XRetriever.retriever import XRDRetriever
from XRetriever.xrd_reader import XRDReader
from XRetriever.peak_detector import PeakDetector

#########################
# INPUT
#########################
# Database path
database_path = 'xrd_database.pkl'

# readin XRD 
import pandas as pd
elements = ['B','Tc'] 
xrd_file = './exp_data/BTc.csv'
xrd_data = pd.read_csv(xrd_file)
two_theta = xrd_data.iloc[:,0]
intensity = xrd_data.iloc[:,1]




#########################
# NO CHANGE
#########################
retriever = XRDRetriever(
    database_path='xrd_database.pkl',
    position_tolerance=0.2,
    min_peak_height=5.0,
    min_peak_prominence=3.0,
    n_peaks=5,
    scoring_method='fom'  # NEW: Use combined scoring
)

detector = PeakDetector(
    min_peak_height=5,
    min_peak_prominence=3.0
)

peaks = detector.get_top_peaks(
    two_theta=two_theta,
    intensity=intensity,
    n_peaks=5,
    preprocess=True
)

exp_positions, exp_intensities = detector.extract_peak_positions_and_intensities(peaks)

results = retriever.retrieve_from_peaks(
    peak_positions=exp_positions.tolist(),
    peak_intensities=exp_intensities.tolist(),
    elements=elements,
    top_n=1
)

entry = retriever.get_entry_details(results[0]['entry_id'])
```
